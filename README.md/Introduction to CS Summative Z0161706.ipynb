a) Write a python function to find the maximum (largest element) of a list. Evaluate the performance of your function by timing its execution on a range of list sizes with the python timeit() function, using a randomly generated a list of numbers as input.

```python
#create a python function to find the maximum of a list,and timing it on a range of list sizes.
#random list one
random.seed(4200)
array1=np.random.randint(100000,size=100)
array1
```

```python
LTOTAL=[L1,L2,L3,L4,L5]
LTOTAL
```

```python
#define a new function
def findmax(M):
    max_value = None
    for i in range(0,len(M)):
        if(max_value is None or M[i] > max_value):
            max_value=M[i]
    return max_value
```

```python
for j in LTOTAL:
    print((findmax(j)))
```

# 3. Finding the upper quartile of a list of numbers.

Write a python function to find the upper quartile of a list, based on the python standard library sorted() function. Compare the performance of this function with the numpy quantile() function and the pandas DataFrame.quantile() function. To find the upper quartile use the quantile function with a q value of 0.75.

```python
#define a python upper quartile function with sorted() function
def up_quartile(u,q):
    upper1 = int(1+(len(u)-1)*q)
    upper2 = 1+(len(u)-1)*q
    sorted_u = sorted(u)
    return sorted_u[upper1-1] + (sorted_u[upper1]-sorted_u[upper1-1])*(upper2-upper1)
```

```python
#find the upper quartile for a single list
up_quartile(u=L1, q=.75)
```

```python
#using for loop to find the upper quartile on a range of list sizes.
for j2 in LTOTAL:
    print(up_quartile(u=j2,q=.75))
```

```python
from timeit import Timer
for j2 in LTOTAL:
    t= Timer(lambda: up_quartile(u=j2,q=.75))
    print(t.timeit(number=1000)/1000)
```

```python
#numpy quantile() function
np.quantile(array1,0.75)
```

```python
for K2 in ATOTAL:
    print(np.quantile(K2,0.75))
```
